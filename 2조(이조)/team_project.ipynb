```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rc('axes', unicode_minus=False)
plt.rcParams['axes.unicode_minus'] = False

file_path = '범죄.csv'  # 업로드된 파일 이름

data = pd.read_csv(file_path, encoding='cp949')  # 인코딩: cp949

# 데이터를 변환 (지역별로 가장 많이 발생한 범죄 중분류 찾기)
regions = [col for col in data.columns if col not in ['범죄대분류', '범죄중분류']]
most_common_crime = data.set_index('범죄중분류')[regions].idxmax()

most_common_crime_df = most_common_crime.reset_index()
most_common_crime_df.columns = ['지역', '가장 많이 발생한 범죄']

most_common_counts = data.set_index('범죄중분류')[regions].max()
most_common_crime_df['발생 건수'] = most_common_counts.values

# 막대 그래프 생성
plt.figure(figsize=(14, 8))  # 그래프 크기 조정
plt.bar(most_common_crime_df['지역'], most_common_crime_df['발생 건수'], color='skyblue')

# 그래프 설정
for i, value in enumerate(most_common_crime_df['발생 건수']):
    plt.text(i, value + 5, most_common_crime_df['가장 많이 발생한 범죄'][i],  # 막대 위에 텍스트 추가
             ha='center', va='bottom', fontsize=10, rotation=45)

plt.title('지역별 가장 많이 발생한 범죄 유형', fontsize=20)  # 제목 크기 증가
plt.xlabel('지역', fontsize=14)  # X축 라벨 크기 증가
plt.ylabel('발생 건수', fontsize=14)  # Y축 라벨 크기 증가
plt.xticks(rotation=45, fontsize=12)  # X축 라벨 회전 및 크기 증가
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rc('axes', unicode_minus=False)
plt.rcParams['axes.unicode_minus'] = False

# 파일 경로
file_path = '범죄.csv'  # 업로드된 파일 이름

# 데이터 로드
data = pd.read_csv(file_path, encoding='cp949')  # 인코딩: cp949

# '강력범죄' 데이터를 필터링
violent_crime_data = data[data['범죄대분류'] == '강력범죄']

# 주요 도시의 데이터 선택 (서울, 부산, 대구, 인천, 광주)
major_cities = ['서울', '부산', '대구', '인천', '광주']
violent_crime_major_cities = violent_crime_data[['범죄중분류'] + major_cities]

# 데이터 시각화 준비
violent_crime_major_cities.set_index('범죄중분류', inplace=True)

# 그래프 생성
ax = violent_crime_major_cities.plot(kind='bar', figsize=(12, 6), width=0.8)

# 막대 위에 이름 표시
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:  # 값이 있는 막대만 표시
            ax.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}',
                    ha='center', va='bottom', fontsize=9)

# 그래프 설정
plt.title('강력범죄 중분류별 주요 도시 발생 건수 비교', fontsize=16)
plt.xlabel('범죄 중분류', fontsize=12)
plt.ylabel('발생 건수', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='도시', fontsize=10)
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rc('axes', unicode_minus=False)
plt.rcParams['axes.unicode_minus'] = False

# 파일 열기
f = open('crime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data) 

# 범죄목 (헤더의 항목들) 추출
crime_categories = header[3:] 

# 서울 지역 데이터 추출
result = []
for row in data:
    if '서울' in row[0]:  # '서울'이 포함된 행만 처리
        for i in row[3:]:
            result.append(int(i))

f.close()

# 그래프 그리기
plt.figure(figsize=(5, 10))
plt.barh(range(len(result)), result, color='blue')

# y축 범죄목 설정
plt.yticks(ticks=range(len(crime_categories)), labels=crime_categories)

# 그래프 설정
plt.xlabel('서울 지역 발생 범죄 수')
plt.ylabel('발생 범죄목')
plt.title('서울 지역 발생 범죄 수')
plt.grid(True, axis='x')  # x축에만 그리드 표시
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python

import csv
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rc('axes', unicode_minus=False)
plt.rcParams['axes.unicode_minus'] = False

# 파일 열기
f = open('crime.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

# 범죄목 (헤더의 항목들) 추출
crime_categories = header[3:]

# 사용자로부터 두 지역 입력받기
region1 = input("첫 번째 지역을 입력하세요: ")
region2 = input("두 번째 지역을 입력하세요: ")

# 각 지역 데이터를 저장할 리스트
region1_result = []
region2_result = []

# 각 지역 데이터를 리스트에 추가
for row in data:
    if region1 in row[0]:  # 첫 번째 지역 데이터 처리
        for i in row[3:]:
            region1_result.append(int(i))
    elif region2 in row[0]:  # 두 번째 지역 데이터 처리
        for i in row[3:]:
            region2_result.append(int(i))

f.close()

# 한 플롯에 두 개의 그래프 생성
plt.figure(figsize=(14, 10))

# 첫 번째 서브플롯: 첫 번째 지역
plt.subplot(1, 2, 1)
plt.barh(range(len(region1_result)), region1_result, color='blue')
plt.yticks(ticks=range(len(crime_categories)), labels=crime_categories)
plt.xlabel(f'{region1} 지역 발생 범죄 수')
plt.ylabel('발생 범죄목')
plt.title(f'{region1} 지역 발생 범죄 수')
plt.grid(True, axis='x')  # x축에만 그리드 표시

# 두 번째 서브플롯: 두 번째 지역
plt.subplot(1, 2, 2)
plt.barh(range(len(region2_result)), region2_result, color='green')
plt.yticks(ticks=range(len(crime_categories)), labels=crime_categories)
plt.xlabel(f'{region2} 지역 발생 범죄 수')
plt.ylabel('발생 범죄목')
plt.title(f'{region2} 지역 발생 범죄 수')
plt.grid(True, axis='x')  # x축에만 그리드 표시

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

file_path = '범죄.csv'
data = pd.read_csv(file_path, encoding='cp949')

seoul_data = data.loc[:, ['범죄대분류', '서울']]

crime_summary = seoul_data.groupby('범죄대분류').sum()

relevant_crimes = crime_summary.loc[['강력범죄', '절도범죄', '폭력범죄', '지능범죄']]

plt.figure(figsize=(8, 8))
plt.pie(relevant_crimes['서울'], labels=relevant_crimes.index, autopct='%1.1f%%', startangle=140)
plt.title('서울의 주요 범죄 비율')
plt.show()
```
